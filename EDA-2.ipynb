```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

#### Unziping file

```python
import gzip
import shutil
with gzip.open('Audience_LA_DMA.csv.gz', 'rb') as f_in:
    with open('Audience_LA_DMA_unzipped.csv', 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)
```

```python
df = pd.read_csv('Audience_LA_DMA_unzipped.csv')
df
```

# Dropping Null Values

```python
df2= df.dropna()
```

```python
df2.shape
```

```python
df.shape
```

#### One Hot encoding
##### We started with Income

```python
df2 = df2.rename(columns = {'ETHNICITY': 'RACE', 'Unnamed: 0':'ID'})
```

```python
df2
```

```python
dummies = pd.get_dummies(df2, columns=['DEMO_HH_INCOME'])
```

```python
dummies.columns
```

```python
one_hot_income = dummies.rename(columns={"DEMO_HH_INCOME_$100,000 to $149,999": "100k - 149k", 
                                                  "DEMO_HH_INCOME_$150,000 to $199,999": "150k - 199k", 
                                                  "DEMO_HH_INCOME_$200,000 to $249,999": "200k - 249k", 
                                                  "DEMO_HH_INCOME_$250,000 +": "250k", 
                                                  "DEMO_HH_INCOME_$30,000 to $49,999": "30k - 49k", 
                                                  "DEMO_HH_INCOME_$50,000 to $74,999": "50k - 74k", 
                                                  "DEMO_HH_INCOME_$75,000 to $99,999": "75k - 99k", 
                                                  "DEMO_HH_INCOME_Less than $30,000": "0 - 30k"})
```

```python
one_hot_income
```

```python
#Shows counts of how many people are in each income bracket

fig_dims = (15, 4)
fig, ax = plt.pyplot.subplots(figsize=fig_dims)

sns.countplot(x="DEMO_HH_INCOME", order = ['Less than $30,000', '$30,000 to $49,999', '$50,000 to $74,999', '$75,000 to $99,999', '$100,000 to $149,999', '$150,000 to $199,999', '$200,000 to $249,999', '$250,000 +'], ax = ax, data=df2)
```

```python
fig_dims = (15, 4)
fig, ax = plt.pyplot.subplots(figsize=fig_dims)
sns.countplot(x="DEMO_HH_INCOME", order = ['Less than $30,000', '$30,000 to $49,999', '$50,000 to $74,999', '$75,000 to $99,999', '$100,000 to $149,999', '$150,000 to $199,999', '$200,000 to $249,999', '$250,000 +'], hue = 'ETHNICITY', hue_order = ['White', 'Asian', 'Hispanic', 'African American'], ax = ax, data=df)
```

```python
#Income by race - percentage. Shows the makeup of each individual income bracket
#Each income bracket adds to 100%

fig_dims = (15, 4)
fig, ax = plt.pyplot.subplots(figsize=fig_dims)

df_plot = df2

x,y = 'DEMO_HH_INCOME', 'RACE'

df_plot = df_plot.groupby(x)[y].value_counts(normalize=True).mul(100).rename('percent').reset_index()

sns.barplot(x=x,y='percent', order = ['Less than $30,000', '$30,000 to $49,999', '$50,000 to $74,999', '$75,000 to $99,999', '$100,000 to $149,999', '$150,000 to $199,999', '$200,000 to $249,999', '$250,000 +'], hue=y, hue_order = ['White', 'Asian', 'Hispanic', 'African American'], ax=ax, data = df_plot)
```

```python
#race by income - shows the percentage of people of a certain race who are in a certain brackat. 
#all the bars for one race add to 100%

fig_dims = (15, 4)
fig, ax = plt.pyplot.subplots(figsize=fig_dims)

df_plot = df

x,y = 'DEMO_HH_INCOME', 'ETHNICITY'

x,y = 'ETHNICITY', 'DEMO_HH_INCOME'


df_plot = df_plot.groupby(x)[y].value_counts(normalize=True).mul(100).rename('percent').reset_index()

sns.barplot(x=y, y='percent', hue=x, order = ['Less than $30,000', '$30,000 to $49,999', '$50,000 to $74,999', '$75,000 to $99,999', '$100,000 to $149,999', '$150,000 to $199,999', '$200,000 to $249,999', '$250,000 +'], hue_order = ['White', 'Asian', 'Hispanic', 'African American'], ax=ax, data = df_plot)
```
