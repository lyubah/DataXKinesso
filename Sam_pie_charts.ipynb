### This function takes any state dataset. It uses the standard column names and plots pie charts representing the proportions of each unique element in the GENDER, DMA_NAME_ACXIOM, DEMO_HH_INCOME, ETHNICITY, HOMEOWNERSHIP_STATUS, and EDUCATION

```python
def show_pie_charts(dataframe):
    
    # DEMO_HH_INCOME
    labels1 = pd.DataFrame(dataframe['DEMO_HH_INCOME'].value_counts(normalize=True)).reset_index().iloc[:, 0]
    sizes1 = pd.DataFrame(dataframe['DEMO_HH_INCOME'].value_counts(normalize=True)).reset_index().iloc[:, 1] * 100
    
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes1, labels=labels1, autopct='%1.1f%%',
        shadow=True, startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    #GENDER
    labels2 = pd.DataFrame(dataframe['GENDER'].value_counts(normalize=True)).reset_index().iloc[:, 0]
    sizes2 = pd.DataFrame(dataframe['GENDER'].value_counts(normalize=True)).reset_index().iloc[:, 1] * 100
    
    fig2, ax2 = plt.subplots()
    ax2.pie(sizes2, labels=labels2, autopct='%1.1f%%',
        shadow=True, startangle=90)
    ax2.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    
    #DMA_NAME_ACXIOM
    labels3 = pd.DataFrame(dataframe['DMA_NAME_ACXIOM'].value_counts(normalize=True)).reset_index().iloc[:, 0]
    sizes3 = pd.DataFrame(dataframe['DMA_NAME_ACXIOM'].value_counts(normalize=True)).reset_index().iloc[:, 1] * 100
    
    fig3, ax3 = plt.subplots()
    ax3.pie(sizes3, labels=labels3, autopct='%1.1f%%',
        shadow=True, startangle=90)  
    ax3.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    
    #ETHNICITY
    labels4 = pd.DataFrame(dataframe['ETHNICITY'].value_counts(normalize=True)).reset_index().iloc[:, 0]
    sizes4 = pd.DataFrame(dataframe['ETHNICITY'].value_counts(normalize=True)).reset_index().iloc[:, 1] * 100
    
    fig4, ax4 = plt.subplots()
    ax4.pie(sizes4, labels=labels4, autopct='%1.1f%%',
        shadow=True, startangle=90)
    ax4.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    
    
    plt.show()
```

```python
show_pie_charts(data_de)
```
